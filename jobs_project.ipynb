```python
# length features
df['desc_len'] = df['Job Description'].str.len()
df['skills_count'] = df['skills'].str.count(',')+1

# topâ€‘words
from collections import Counter
words = Counter(" ".join(df['skills']).split())
words.most_common(20)
```

```python
df['Experience'].value_counts()
```

```python
df['Salary Range'] = df['Salary Range'].str.replace('$', '').str.replace('K', '')
df[['salary_min', 'salary_max']] = df['Salary Range'].str.split('-', expand=True)

df['salary_min'] = df['salary_min'].astype(float) * 1000
df['salary_max'] = df['salary_max'].astype(float) * 1000

df['salary_avg'] = df[['salary_min', 'salary_max']].mean(axis=1)
```

```python
sns.histplot(df['salary_avg'], bins=30)
plt.title('Distribution of Average Salary')
plt.xlabel('Salary ($)')
plt.ylabel('Frequency')
plt.show()
```
